```python
words = open('names.txt', 'r').read().splitlines()
words[:5]
```

```python
b = {}
for w in words:
    chs = ['<S>'] + list(w) + ['<E>']
    for ch1,ch2 in zip(chs, chs[1:]):
        bigram = (ch1,ch2)
        b[bigram] = b.get(bigram, 0) + 1
```

```python
sorted(b.items(), key = lambda x: x[1], reverse = True)
```
